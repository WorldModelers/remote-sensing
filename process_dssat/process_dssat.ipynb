```python
import fiona
from shapely.geometry import Point, shape
import csv

# Match longitude/latitude to a specific district
# 
# 
def long_lat_to_district(lng, lat, shape_file):
    """ Match longitude/latitude to a specific district within a shape file
    
    Args:
        lng (float): Longitude
        lat (float): Latitude
        shape_file (string): Path to shape file
    
    Returns:
        object: Returns properties associated with the district provided
                latitude/longitude is positioned within
    """
    mypoint = Point(lng, lat) 

    with fiona.open(shape_file) as shp:
        polygons = [poly for poly in shp]

    polygon_ids = [i for i, poly in enumerate(polygons)
                if mypoint.within(shape(poly['geometry']))]

    if not polygon_ids:
        return None
    else:
        # Take first polygon that overlaps since may overlap with several if on border
        match = polygons[polygon_ids[0]]
        return match['properties']
    
def append_districts_to_dssat(dssat_file, shape_file, out_file):
    """ Append district metadata to dssat and writes to outfile
    
    Args:
        dssat_file: Path to dssat file
        shape_file: Path to shape file
        
    Return:
        None
    """
    wf = open(out_file, 'w', newline='')
    writer = csv.writer(wf)

    rf = open(dssat_file, 'r')
    reader = csv.reader(rf)
    header = next(reader, None)
    writer.writerow(header + ['NAME_1', 'NAME_2', 'NAME_3', 'CC_3'])
    
    for index, row in enumerate(reader):
        if index % 100 == 0:
            print(index)
        lat = float(row[0])
        lng = float(row[1])
        district_props = long_lat_to_district(lng, lat, shape_file)
        writer.writerow(row + [district_props['NAME_1'], district_props['NAME_2'] , district_props['NAME_3'], district_props['CC_3']])
        
    rf.close()
    wf.close()
    
```
